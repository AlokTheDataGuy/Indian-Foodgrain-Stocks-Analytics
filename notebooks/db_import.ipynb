```python
paddy_df = pd.read_csv('../data/raw/paddy_2011_2025.csv')
rice_parboiled_df = pd.read_csv('../data/raw/rice_parboiled_2011_2025.csv')
rice_raw_df = pd.read_csv('../data/raw/rice_raw_2011_2025.csv')
wheat_df = pd.read_csv('../data/raw/wheat_2011_2025.csv')
```

```python
# Connecting to MySQL
from sqlalchemy import create_engine

username = "pyuser"
password = "py_password"
host = "localhost"
port = "3306"
database = "foodgrain_stocks"

engine = create_engine(
    f"mysql+pymysql://{username}:{password}@{host}:{port}/{database}"
)

# Loading Dataframe into MySQL
table_name = "paddy_raw"
paddy_df.to_sql(table_name, engine, index=False)

print(f"Data successfully loaded into table '{table_name}' in database '{database}'.")

table_name = "rice_parboiled_raw"
rice_parboiled_df.to_sql(table_name, engine, index=False)

print(f"Data successfully loaded into table '{table_name}' in database '{database}'.")

table_name = "rice_raw_raw"
rice_raw_df.to_sql(table_name, engine, index=False)

print(f"Data successfully loaded into table '{table_name}' in database '{database}'.")

table_name = "wheat_raw"
wheat_df.to_sql(table_name, engine, index=False)

print(f"Data successfully loaded into table '{table_name}' in database '{database}'.")
```
